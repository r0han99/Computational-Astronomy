### 1.Splitting the train and test sets

```python
import numpy as np

def splitdata_train_test(data, fraction_training):
    np.random.seed(0)
    np.random.shuffle(data)
    split = int(len(data)*fraction_training)
    trainings_set = data[:split]
    testing_set = data[split:]
    
    return trainings_set, testing_set
    
  
  

if __name__ == "__main__":
  data = np.load('galaxy_catalogue.npy')

  
  fraction_training = 0.7

  
  training, testing = splitdata_train_test(data, fraction_training)


  print('Number data galaxies:', len(data))
  print('Train fraction:', fraction_training)
  print('Number of galaxies in training set:', len(training))
  print('Number of galaxies in testing set:', len(testing))
```

### Generating features and targets

```python
import numpy as np

def generate_features_targets(data):


  targets = data['class']

  features = np.empty(shape=(len(data), 13))
  features[:, 0] = data['u-g']
  features[:, 1] = data['g-r']
  features[:, 2] = data['r-i']
  features[:, 3] = data['i-z']
  features[:, 4] = data['ecc']
  features[:, 5] = data['m4_u']
  features[:, 6] = data['m4_g']
  features[:, 7] = data['m4_r']
  features[:, 8] = data['m4_i']
  features[:, 9] = data['m4_z']


  # concentration in u filter
  features[:, 10] = data['petroR50_u']/data['petroR90_u']
  # concentration in r filter
  features[:, 11] = data['petroR50_r']/data['petroR90_r']
  # concentration in z filter
  features[:, 12] = data['petroR50_z']/data['petroR90_z']

  return features, targets


if __name__ == "__main__":
  data = np.load('galaxy_catalogue.npy')

  features, targets = generate_features_targets(data)

  print("Features shape:", features.shape)
  print("Targets shape:", targets.shape)
```

### Train the decision tree classifier

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report


def splitdata_train_test(data, fraction_training):
    np.random.seed(0)
    np.random.shuffle(data)
    split = int(len(data)*fraction_training)
    trainings_set = data[:split]
    testing_set = data[split:]
    
    return trainings_set, testing_set


def generate_features_targets(data):


  targets = data['class']

  features = np.empty(shape=(len(data), 13))
  features[:, 0] = data['u-g']
  features[:, 1] = data['g-r']
  features[:, 2] = data['r-i']
  features[:, 3] = data['i-z']
  features[:, 4] = data['ecc']
  features[:, 5] = data['m4_u']
  features[:, 6] = data['m4_g']
  features[:, 7] = data['m4_r']
  features[:, 8] = data['m4_i']
  features[:, 9] = data['m4_z']


  # concentration in u filter
  features[:, 10] = data['petroR50_u']/data['petroR90_u']
  # concentration in r filter
  features[:, 11] = data['petroR50_r']/data['petroR90_r']
  # concentration in z filter
  features[:, 12] = data['petroR50_z']/data['petroR90_z']

  return features, targets



def dtc_predict_actual(data):

    training_set, testing_set = splitdata_train_test(data,0.7)



    train_features, train_targets = generate_features_targets(training_set)
    test_features, test_targets = generate_features_targets(testing_set)


    dtr = DecisionTreeClassifier()

    dtr.fit(train_features,train_targets)

    predictions = dtr.predict(test_features)

    return predictions, test_targets



if __name__ == '__main__':
  data = np.load('galaxy_catalogue.npy')
    
  predicted_class, actual_class = dtc_predict_actual(data)

  # Print some of the initial results
  print("Some initial results...\n   predicted,  actual")
  for i in range(10):
    print("{}. {}, {}".format(i, predicted_class[i], actual_class[i]))
 
```

### Accuracy in classification

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
style.use('dark_background')
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.tree import DecisionTreeClassifier
from support_functions import plot_confusion_matrix, generate_features_targets



def calculate_accuracy(predicted, actual):
    
    correct_preds = sum(predicted == actual)
    
    accuracy =  correct_preds / len(actual)
    
    return accuracy
    
    
    
      
if __name__ == "__main__":
  data = np.load('galaxy_catalogue.npy')


  features, targets = generate_features_targets(data)


  dtc = DecisionTreeClassifier()
  dtr_predicted = cross_val_predict(dtc, features, targets, cv=10)


  model_score = calculate_accuracy(dtr_predicted, targets)
  print("Our accuracy score: {:.2f}".format(model_score))


  class_labels = list(set(targets))
  model_cm = confusion_matrix(y_true=targets, y_pred=dtr_predicted, labels=class_labels)

  # Plot the confusion matrix using the provided functions.
  plt.figure(figsize=(7,7))
  plot_confusion_matrix(model_cm, classes=class_labels, normalize=False)
  plt.show()
```

### Random Forest

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
style.use('dark_background')
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from support_functions import generate_features_targets, plot_confusion_matrix


def rf_predict_actual(data, n_estimators):

    features, targets = generate_features_targets(data)

    rfc = RandomForestClassifier(n_estimators)

    rfpredicted = cross_val_predict(rfc, features, targets,cv=10)

    return rfpredicted, targets

if __name__ == "__main__":
  data = np.load('galaxy_catalogue.npy')


  number_estimators = 50              # Number of trees
  rfpredicted, actual = rf_predict_actual(data, number_estimators)


  accuracy = calculate_accuracy(rfpredicted, actual)
  print("Accuracy score:", accuracy)


  class_labels = list(set(actual))
  model_cm = confusion_matrix(y_true=actual, y_pred=rfpredicted, labels=class_labels)

  plt.figure(figsize=(7,7))
  plot_confusion_matrix(model_cm, classes=class_labels, normalize=False)
  plt.show()
```

### Elbow Techinque Choosing K-value

```python
error_rate = []
for k in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_features,train_targets)
    knn_pred_k = knn.predict(test_features)
    error_rate.append(np.mean(knn_pred_k != test_targets))
    
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,50),error_rate,color='r',marker='o')
```

```python
import cufflinks as cf 
import plotly.graph_objects as go
cf.go_offline()
```

```python
x = list(range(1,50))
fig = go.Figure( 
    layout=go.Layout(
        title=go.layout.Title(text="Elbow Technique ( Error Rate Vs K-value)")))
    
fig.add_trace(go.Scatter(x=x, y=error_rate, name = 'Error Rate v K-value',mode='lines+markers',
                         line=dict(color='black', width=2)))
fig.show()
```

### Creating a Neural Network 

```python
training_set, testing_set = splitdata_train_test(data,0.7)
train_features, train_targets = generate_features_targets(training_set)
test_features, test_targets = generate_features_targets(testing_set)
```

```python
train_targets = pd.Series(train_targets).map({'merger':2,'elliptical':1,'spiral':0})
train_targets = np.array(train_targets)
test_targets = pd.Series(test_targets).map({'merger':2,'elliptical':1,'spiral':0})
test_targets = np.array(test_targets)
```

```python
# normalising the data 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
train_features.shape
```

```python
# instantiation
model = Sequential()

model.add(Dense(13,activation='relu'))


model.add(Dense(1,activation='softmax'))

model.compile(loss='categorical_crossentropy', 
                      optimizer='adam', 
                      metrics=['accuracy'])
```

```python
import time
start = time.perf_counter()
# fitting the model
model.fit(x=train_features,y=train_targets,epochs=200)
# epochs selected arbitraryly 

print("\nTIME ELAPSED {}".format(time.perf_counter() - start))
```
